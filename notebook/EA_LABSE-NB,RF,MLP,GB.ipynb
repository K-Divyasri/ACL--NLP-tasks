```python
import string
def removepunctuation(text):
    clean="".join([i for i in text if i not in string.punctuation])
    return clean

#df['text']= df['text'].apply(lambda x:removepunctuation(x))
#df.head()
```

```python
import re
def deEmojify(text):
    regrex_pattern = re.compile(pattern = "["
        u"\U0001F600-\U0001F64F"  # emoticons
        u"\U0001F300-\U0001F5FF"  # symbols & pictographs
        u"\U0001F680-\U0001F6FF"  # transport & map symbols
        u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                           "]+", flags = re.UNICODE)
    return regrex_pattern.sub(r'',text)

#df['text']= df['text'].apply(lambda x:deEmojify(x))
#df.iloc[12]
```

```python
with open('tamil_stopwords.txt', encoding='utf-8') as file:
    stopwords = file.read().splitlines()

def removestopwords(text):
    removed= " ".join([s for s in text.split() if s not in stopwords])
    return removed
```

```python
dictkeys = {'Anger': 0, 'Neutral': 1, 'Joy': 2, 'Disguist': 3, 'Trust': 4, 'Anticipation': 5,
       'Ambiguous':6, 'Love': 7, 'Surprise':8, 'Sadness': 9, 'Fear':10}
#df["emokeys"] = df["emo"].map(dictkeys)
#df
```

```python
def preprocess(df):
    df['text']= df['text'].apply(lambda x:removepunctuation(x))
    df['text']= df['text'].apply(lambda x:deEmojify(x))
    df['text_wo_sw']= df['text'].apply(lambda x:removestopwords(x))
    df["emokeys"] = df["emo"].map(dictkeys)
    return df
```

```python
df = preprocess(df)
df_fin = df[["text_wo_sw", "emokeys"]]
```
